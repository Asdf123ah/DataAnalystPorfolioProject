# Scraping Data from a Real Website + Pandas

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
<table class="wiketable sortable">
<caption>
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

revenueperworker_index = world_table_titles.index('Headquarters[note 1]')

limited_titles = world_table_titles[:revenueperworker_index + 1]

print(limited_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = limited_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[2:]:
    row_data = row.find_all(['th','td'])
    individual_row_data = [data.text.strip() for data in row_data][:7]
#   print(individual_row_data)
    length = len(df) 
    df.loc[length] = individual_row_data
```
