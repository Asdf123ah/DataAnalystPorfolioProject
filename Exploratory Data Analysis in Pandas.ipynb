```python
# Select only numeric columns from the DataFrame before calculating correlation
numeric_df = df.select_dtypes(include=['number'])  # This will include only numeric columns
correlation_matrix = numeric_df.corr()  # Now calculate the correlation
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot = True)

plt.rcParams['figure.figsize'] = (20, 10)
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```
